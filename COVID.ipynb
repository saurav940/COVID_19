```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('StatewiseTestingDetails.csv')
print(df)
```

```python
import matplotlib.pyplot as plt
x=df['TotalSamples']
y=df['Negative']
plt.plot(x,y)
plt.ylabel("Negative")
plt.xlabel("Total sample")
plt.title('Direction ')
```

```python
import matplotlib.pyplot as pltpie
df['Positive'].plot(kind='pie',
                            figsize=(20,18),
                            autopct='%1.1f%%', 
                            startangle=45,     
                 
                            
                            )

pltpie.title('positive case')
pltpie.axis('equal') # Sets the pie chart to look like a circle.
pltpie.show()
```

```python
import pandas as pd
StatewiseTestingDetails=pd.read_csv('StatewiseTestingDetails.csv')
StatewiseTestingDetails
```

```python
import matplotlib.pyplot as plt
StatewiseTestingDetails['Positive'].plot(kind='pie',
                            figsize=(19,8),
                            autopct='%1.1f%%', 
                            startangle=60,  
                            pctdistance=1.12,
                            labels=None,   
                            )
plt.title('Total Positive Cases according to State')
plt.legend(labels=StatewiseTestingDetails.index, loc='upper left') 
plt.axis('equal')
plt.show()
```
